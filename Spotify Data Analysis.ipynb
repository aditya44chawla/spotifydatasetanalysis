### Visualizing scatter plot of danceability and loudness and danceability and energy

```python
fig, ax =plt.subplots(2,1, figsize=(15,20))
i = sns.scatterplot(x = df['loudness'], y = df['danceability'], ax = ax[0], color = "blue")
j = sns.scatterplot(x = df['energy'], y = df['danceability'], ax = ax[1], color = "orange")
i.set_title("Relationship between danceability and loudness")
j.set_title("Relationship between danceability and energy")
```

### Importing libraries for using heat map

```python
%matplotlib inline
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

### Using heat map to see th relationship between danceability, energy, loudness, popularity, duration and explicit

## Figure 3

```python
# Defining correlation between different variables using heatmap
corr_mat = df.corr()
plt.figure(figsize = (15, 15))
ax = plt.axes()
sns.heatmap(corr_mat,
            annot=True,
            fmt='.2f',
            cmap=plt.cm.coolwarm_r, ax=ax
           )
ax.set_title('Relationship between 6 variables')
plt.show()
```
